````python
def get_mysql_coneection(HOST, USER, PASSWORD, DATABASE):
    try:
        # MySQL connection details
        connection = mysql.connector.connect(
            host=HOST,           # or your host
            user=USER,       # replace with your MySQL username
            password=PASSWORD,   # replace with your password
            database=DATABASE    # replace with your database name
        )
        return connection
    except Exception as e:
        raise e
````

### Function execute queries

````python
def execute_query(
    my_query: str = "SELECT * FROM gold_dim_customers;",
    host: str = HOST,
    user: str = USER,
    password: str = PASSWORD,
    database: str = DATABASE
) -> pd.DataFrame:
    try:
        # Connect to MySQL database
        conn = mysql.connector.connect(
            host=host,
            user=user,
            password=password,
            database=database
        )

        # Create a cursor object
        cursor = conn.cursor()

        # Execute the SQL command
        cursor.execute(my_query)

        # Fetch all results
        results = cursor.fetchall()

        # Convert results to pandas DataFrame
        df = pd.DataFrame(results, columns=[column[0] for column in cursor.description])
        # print(df)

        # Close the connection
        cursor.close()
        conn.close()

        return df

    except Exception as e:
        print(f"Error: {e}")
        return pd.DataFrame()
````

```sql

SELECT
```

```sql
LEFT JOIN gold_dim_products p
    ON p.product_key = f.product_key
GROUP BY p.product_name
```

````python
execute_query(
    """ 
    SELECT
        p.product_name,
        SUM(f.sales_amount) AS total_revenue
    FROM gold_fact_sales f
    LEFT JOIN gold_dim_products p
        ON p.product_key = f.product_key
    GROUP BY p.product_name
    ORDER BY total_revenue DESC
    LIMIT 5;

    """
)
````

```sql

-- What are the 5 worst-performing products in terms of sales ?

-- Solution without window function

SELECT TOP 5
p.subcategory,
SUM(f.sales_amount) AS total_revenue
FROM gold_fact_sales f
LEFT JOIN gold_dim_products p
ON p.product_key = f.product_key
GROUP BY p.subcategory
ORDER BY total_revenue DESC;

-- Solution with window function
SELECT *
FROM (
    SELECT TOP 5
    p.product_name,
    SUM(f.sales_amount) AS total_revenue,
    ROW_NUMBER() OVER (ORDER BY SUM(f.sales_amount) DESC) AS rank_products
    FROM gold_fact_sales f
    LEFT JOIN gold_dim_products p
    ON p.product_key = f.product_key
    GROUP BY p.product_name)t
WHERE rank_products <= 5
ORDER BY total_revenue;

```

````python
# without window function
execute_query(
    """ 
    SELECT
        p.subcategory,
        SUM(f.sales_amount) AS total_revenue
    FROM gold_fact_sales f
    LEFT JOIN gold_dim_products p
        ON p.product_key = f.product_key
    GROUP BY p.subcategory
    ORDER BY total_revenue DESC
    LIMIT 5;
    """
)
````

````python
# with window function
execute_query(
    """ 
    SELECT *
    FROM (
        SELECT
            p.product_name,
            SUM(f.sales_amount) AS total_revenue,
            ROW_NUMBER() OVER (ORDER BY SUM(f.sales_amount) DESC) AS rank_products
        FROM gold_fact_sales f
        LEFT JOIN gold_dim_products p
            ON p.product_key = f.product_key
        GROUP BY p.product_name
    ) t
    WHERE rank_products <= 5
    ORDER BY total_revenue DESC;

    """
)
````
